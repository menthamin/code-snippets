```python
import pandas as pd

df = pd.read_csv("data/분석1.csv")
df.rename(columns={"Unnamed: 0": "id"}, inplace=True)
```

# 1. 매핑용 사전 만들기

```python
성과유형_list = []
성과구분_list = []
성과구분_CD_list = []

for col_nm in df.columns:
    if col_nm.startswith("성과유형"):
        성과유형_list.append(col_nm)
    elif col_nm.startswith("성과구분"):
        if col_nm.startswith("성과구분_CD"):
            성과구분_CD_list.append(col_nm)
        else:
            성과구분_list.append(col_nm)
```

# 2. 계산

## 2.1. 계산

```python
sum_list = []

for idx, row in df.iterrows():
    long_sum = 0
    long_cnt = 0
    short_sum = 0
    short_cnt = 0
    for key, value in row.items():
        if key in 성과유형_구분_CD_dict:
            if value == "장기":
                성과구분_value = row[성과유형_구분_CD_dict[key]]
                long_sum += 성과구분_value
                long_cnt += 1
            elif value == "단기":
                성과구분_value = row[성과유형_구분_CD_dict[key]]
                short_sum += 성과구분_value
                short_cnt += 1
    
    sum_list.append([long_sum, long_cnt, short_sum, short_cnt])
```
