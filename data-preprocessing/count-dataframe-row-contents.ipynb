# 방법 1: 조건문 if 활용

```python
df = pd.read_csv("data/count_sample.csv")
```

```python
count_long_list = []
count_short_list = []

for row in df.iterrows():
    # row_num = row[0]
    row_data = list(row[1])
    count_long_case = 0
    count_short_case = 0

    for data in row_data[0:5]:
        if data == "장기":
            count_long_case += 1
        elif data == "단기":
            count_short_case += 1

    count_long_list.append(count_long_case)
    count_short_list.append(count_short_case)

df["장기개수"] = count_long_list
df["단기개수"] = count_short_list
```

# 방법 2: Counter 함수 사용
- 카운트 해야하는 데이터의 종류가 다양할 때 유용
- 장기, 단기 처럼 2개 케이스만 카운트하는 경우에는 if문과 별차이 없음

```python
from collections import Counter
```

```python
df = pd.read_csv("data/count_sample.csv")
df
```

```python
count_long_list = []
count_short_list = []

for row in df.iterrows():
    # row_num = row[0]
    row_data = list(row[1])
    count_result = Counter(row_data[0:5])  # Counter
    print(row_data)
    print(count_result)

    count_long_case = count_result["장기"]
    count_short_case = count_result["단기"]

    count_long_list.append(count_long_case)
    count_short_list.append(count_short_case)

df["장기개수"] = count_long_list
df["단기개수"] = count_short_list
```
